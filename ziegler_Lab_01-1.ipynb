```python
# Above Average
running_average = 0
total_count = 0
count = 0

for i in bmi_value:
    running_average += i
    total_count += 1
average = running_average / total_count

above_average = [a for a in bmi_value if a > average]
for n in above_average:
    count += 1
print("The average bmi in this sample is:",average)
print("The following number of data points are above the average:",count)
print(above_average)
```

```python
# Below Average

running_average = 0
total_count = 0
count = 0

for i in bmi_value:
    running_average += i
    total_count += 1
average = running_average / total_count

below_average = [a for a in bmi_value if a < average]
for n in below_average:
    count += 1
print("The average bmi in this sample is:",average)
print("The following number of data points are above the average:",count)
print(below_average)
```
